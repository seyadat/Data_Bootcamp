```python
#IMPORTING IN THE FRAME WORKS

from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://capital.com/lp-trade-ga-bah?utm_medium=cpc&utm_source=google&utm_campaign=web_bah_search_google_me_en_tcpa_brand-ftd_ua&utm_term=capital%20best%20trading%20platform&adgroupid=174860264144&matchtype=e&creative=745438932675&device=c&loc_physical=1011785&gad_source=1&gad_campaignid=21759029956&gbraid=0AAAAADQE_svPY04OInPJFh5JRCQken3Tk&gclid=Cj0KCQiAosrJBhD0ARIsAHebCNoxgr5oq6y7Zf9u9QzK8w47vd7dGcVQjHSPVitcnEDAC6untAZN04QaAi8aEALw_wcB'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}
page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup.find_all('table')
```

```python
table = soup.find('table', class_="table table--fixedInstr table--interactive")

headers = table.find_all('th')
```

```python
header_title = [header.text.strip() for header in headers]

print(header_title)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=header_title)
```

```python
df
```

```python
column_row = table.find_all('tr')
```

```python
for row in column_row[1:]:
    row_data=row.find_all('td')
    individual_data = [data.text.strip() for data in row_data]

    length = len(df)  
    df.loc[length] = individual_data
```

```python
df
```

```python
df = df.rename(columns={'ame': 'Name'})
```
