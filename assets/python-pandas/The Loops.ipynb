```python
#tow types of loops
    #while loops-> can execute a set of statement as long as a condition is true
    #for loops -> it loops or iterate over a sequence(can be a list, tuple, dictionary, a set or a string)
```

```python
i = 2
while i < 10:
    print(i)
    i += 7
```

```python
j = 3
while j < 14:
    print(j)
    if j == 9:
        break
    j += 3
```

```python
number = 1

while number < 10:
    print(number)
    number = number+3
```

```python
#break statement
```

```python
number = 1

while number < 10:
    print(number)
    if number == 7:
        break
    number = number+3
```

```python
number = 1

while number < 10:
    print(number)
    if number == 7:
        break
    number = number+3
else:
    print('no longer available')
```

```python
#continue statement 
```

```python
number = 1

while number < 10:
      number = number+3
    if number == 6:
            continue
        print(number)
else:
    print('no longer available')
```

```python
#for loops
    #does not require an indexing variable to set beforehand
```

```python
days = [1, 2, 5, 6, 9]

for x in days:
    print('yeah!')
```

```python
days = [1, 2, 5, 6, 9]

for x in days:
    print(x * 10)
```

```python
fruits = ['apple', 'banana', 'cherry']
for x in fruits:
    print(x)
```

```python
Market_list = {'fruit':['mango', 'Pawpaw', 'Pineapple'], 'Cleaning_ware':['soap', 'washpower', 'sponge'], 'day': 6, 'ice_flavor':['chocolate', 'vanilla']}
for x in Market_list.values():
    print(x)
```

```python
Market_list = {'fruit':['mango', 'Pawpaw', 'Pineapple'], 'Cleaning_ware':['soap', 'washpower', 'sponge'], 'day': 6, 'ice_flavor':['chocolate', 'vanilla']}
for x in Market_list.items():
    print(x)
```

```python
Market_list = {'fruit':['mango', 'Pawpaw', 'Pineapple'], 'Cleaning_ware':['soap', 'washpower', 'sponge'], 'day': 6, 'ice_flavor':['chocolate', 'vanilla']}
for keys, values in Market_list.items():
    print(keys,'->', values)
```

```python
#nested for loops
```

```python
abst = ['red', 'big', 'tasty']
fruits = ["apple", "banana", "cherry"]

for one in abst:
    for two in fruits:
        print(one, 'matches', fruits)
```

```python
#nested for loops
```

```python
flavor = ['vanilla', 'chocolate', 'strawberry']
toppings = ['Biscuits', 'cornflakes', 'sherims']
```

```python
for x in flavor:
    for y in toppings:
        print(y, 'topped with', y)
```
