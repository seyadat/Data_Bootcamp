# Merge Func()

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\Seyy_\OneDrive\Desktop\jupyter projects\alex the analyst bootcamp\python\employees.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\Seyy_\OneDrive\Desktop\jupyter projects\alex the analyst bootcamp\python\salaries.csv")
df2
```

```python
#using .merge()

df1.merge(df2)
```

```python
pd.merge(df1, df2, on='employee_id')
```

```python
#specifying the specific merge, we use Inner  Merger how ='' , on =''
```

```python
pd.merge(df1, df2, how = 'inner', on='employee_id')
```

```python
#Outer Join
```

```python
pd.merge(df1, df2, how = 'outer', on='employee_id')
```

```python
#Lefter Merge
```

```python
pd.merge(df1, df2, how = 'left', on='employee_id')
```

```python
#Right Marge
```

```python
pd.merge(df1, df2, how = 'right', on='employee_id')
```

```python
#Cross Merge,
#=> compares values from the left with each value the right on at a time
```

```python
pd.merge(df1, df2, how = 'cross')
```

# Join func()

```python
#
```

```python
df1.join(df2, how = 'outer', on = 'employee_id', lsuffix='_left', rsuffix='_right')
```

```python
 #to set the index right and Join correctly
#set the index the pri & foreign key values
```

```python
df3 = df1.set_index('employee_id').join(df2.set_index('employee_id'), how='outer', on='employee_id', lsuffix='_left', rsuffix='_right' )
df3
```

```python
#inner join
df1.set_index('employee_id').join(df2.set_index('employee_id'), how='inner', on='employee_id', lsuffix='_left', rsuffix='_right' )
```
