```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_telecommunications_companies'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115 Safari/537.36'
}
```

```python
response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table = soup.find('table', class_='wikitable sortable')
```

```python
header = table.find_all('th')
```

```python
head_title = [title.text.strip() for title in header]

print(head_title)
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(columns=head_title)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    insert_data = [data.text.strip() for data in row_data]

    df.loc[len(df)] = insert_data
```

```python
df
```

```python
df = df.set_index('Rank')
```
