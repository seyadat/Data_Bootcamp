```python
calendar_days = {1, 3, 101, 59, 13, 3, 5, 19, 39, 20, 1}

print(calendar_days)
#display the unique values in the set
```

```python
holiday_days ={13, 103, 1, 40, 19, 59, 300, 3, 12, 6}
```

```python
# | the bar gives the combines unique of 2 compared sets
# & gives values that show up in both set
# - gives values that does show on both set
# ^ show values either in set 1 or set 2 and not in both

print( calendar_days | holiday_days)

print(calendar_days & holiday_days)

print(calendar_days - holiday_days)

print(calendar_days ^ holiday_days)
```
