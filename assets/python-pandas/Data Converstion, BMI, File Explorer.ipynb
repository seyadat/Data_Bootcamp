# BMI CALCULATOR

```python
#trying out the BMI calculator trials
```

```python
name = input('Enter your name!')

age = int(input('Enter your age'))

weight = int(input('Enter weight in pounds:'))

height = int(input('Enter your height:'))

BMI = (weight * 703) / (height * height)

print(BMI)

if BMI > 0:
    if (BMI < 16):
        print('Dear ', name + ' your BMI statues is severe Thinness')
    elif BMI < 17:
        print('Dear ', name + ' your BMI statues is Moderate Thinness')
    elif BMI < 18.8:
        print('Dear ', name + ' your BMI statues is Mild Thinness')
    elif BMI < 24.9:
        print('Dear ', name + ' your BMI statues is Normal')
    elif BMI < 29.9:
        print('Dear ', name + ' your BMI statues is Overweight')
    elif BMI < 34.9:
        print('Dear ', name + ' your BMI statues is Obese Class I')
    elif (BMI < 39.9):
        print('Dear ', name + ' your BMI statues is Obese Class II')
    else:
        print('Dear ', name + ' your BMI statues is Obese Class III') 
else:
    print('Enter an invalid input!')
```

```python
Severe Thinness	< 16
Moderate Thinness	16 - 17
Mild Thinness	17 - 18.5
Normal	18.5 - 25
Overweight	25 - 30
Obese Class I	30 - 35
Obese Class II	35 - 40
Obese Class III	> 40
```

# AUTOMATIC FILE SORTER IN FILE EXPLORER

```python
.exe, .csv, .zip, .pdf, .docx
```

```python
import os, shutil
```

```python
path = r"C:/Users\Seyy_/Downloads/"
```

```python
file_name = os.listdir(path)
```

```python
folder_names = ['zip folder', 'pdf folder','csv folder', 'setup folder']

for loop in range(0,4):
    if not os.path.exists(path + folder_names[loop]):
        print(path + folder_names[loop])
        os.makedirs(path + folder_names[loop])
```

```python
for file in file_name:
    if '.csv' in file and not os.path.exists(path + "csv folder/" + file):
        shutil.move(path + file, path+ "csv folder/" + file )
    elif '.zip' in file and not os.path.exists(path + "zip folder/" + file):
        shutil.move(path + file, path+ "zip folder/" + file )
    elif '.pdf' in file and not os.path.exists(path + "pdf folder/" + file):
        shutil.move(path + file, path+ "pdf folder/" + file )
    elif '.setup' in file and not os.path.exists(path + "setup folder/" + file):
        shutil.move(path + file, path+ "setup folder/" + file )
else:
    print('file path were not found')
```
